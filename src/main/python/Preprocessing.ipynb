```python
import pandas as pd
from sklearn import preprocessing
```

```python
df_raw = pd.read_csv("C:\\Users\\ADMIN\\Desktop\\balance_open_timeseries.csv", header = None, low_memory=False)
```

```python
df = df_raw.transpose().sample(n = 5000)
df = df.transpose()
for i in range(45):
    df.iloc[i:i+15].transpose().to_csv("C:\\Users\\ADMIN\\Documents\\raw_data_5k\\month_" + str(i) + ".csv", header = False)
    print(i)
```

```python
df = df_raw.iloc[0:12]
```

```python
x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
```

```python
df = df.drop(df.columns[df.sum() == 0], axis = 1)
df = df.transpose()
```
